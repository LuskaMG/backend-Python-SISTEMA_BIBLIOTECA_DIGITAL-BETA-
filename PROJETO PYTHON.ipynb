```python
lista_livros      = []
lista_usuarios    = []
lista_emprestimos = []

proximo_id_livro      = 1
proximo_id_usuario    = 1
proximo_id_emprestimo = 1


def buscar_livro_por_id(livro_id):
    for livro in lista_livros:
        if livro["id"] == livro_id:
            return livro
    return None


def buscar_usuario_por_id(usuario_id):
    for usuario in lista_usuarios:
        if usuario["id"] == usuario_id:
            return usuario
    return None


def buscar_emprestimo_por_livro(livro_id):
    for emp in lista_emprestimos:
        if emp["livro_id"] == livro_id:
            return emp
    return None


def separador(char="─", largura=72):
    print("  " + char * largura)


def inicializar_dados():
    global proximo_id_livro

    livros_iniciais = [
        ("Dom Casmurro",                    "Machado de Assis",       "Romance"),
        ("Memórias Póstumas de Brás Cubas", "Machado de Assis",       "Romance"),
        ("O Cortiço",                       "Aluísio Azevedo",        "Naturalismo"),
        ("Grande Sertão: Veredas",          "Guimarães Rosa",         "Romance"),
        ("Capitães da Areia",               "Jorge Amado",            "Romance"),
        ("A Hora da Estrela",               "Clarice Lispector",      "Romance"),
        ("Vidas Secas",                     "Graciliano Ramos",       "Romance"),
        ("1984",                            "George Orwell",          "Ficção Científica"),
        ("O Senhor dos Anéis",              "J.R.R. Tolkien",         "Fantasia"),
        ("Cem Anos de Solidão",             "Gabriel García Márquez", "Realismo Mágico"),
        ("Crime e Castigo",                 "Fiódor Dostoiévski",     "Romance"),
        ("O Pequeno Príncipe",              "Antoine de Saint-Exupéry","Fábula"),
        ("Sapiens",                         "Yuval Noah Harari",      "História"),
        ("O Andar do Bêbado",               "Leonard Mlodinow",       "Ciência"),
        ("Código Limpo",                    "Robert C. Martin",       "Tecnologia"),
    ]

    for titulo, autor, genero in livros_iniciais:
        lista_livros.append({
            "id": proximo_id_livro,
            "titulo": titulo,
            "autor": autor,
            "genero": genero,
            "disponivel": True,
        })
        proximo_id_livro += 1

    print("\n  ✔  Sistema iniciado com 15 livros cadastrados.\n")


def cadastrar_livro():
    global proximo_id_livro
    print("\n┌─── Cadastrar Livro ──────────────────────────────────")
    titulo = input("│  Título  : ").strip()
    autor  = input("│  Autor   : ").strip()
    genero = input("│  Gênero  : ").strip()

    if not titulo or not autor:
        print("└  ✘  Título e autor são obrigatórios.\n")
        return

    lista_livros.append({
        "id": proximo_id_livro,
        "titulo": titulo,
        "autor": autor,
        "genero": genero if genero else "Não informado",
        "disponivel": True,
    })
    print(f"└  ✔  Livro '{titulo}' cadastrado com ID {proximo_id_livro}.\n")
    proximo_id_livro += 1


def listar_livros():
    print("\n┌─── Acervo Completo ──────────────────────────────────────────────────────")

    if not lista_livros:
        print("│  Nenhum livro cadastrado.\n└")
        return

    disponiveis = [l for l in lista_livros if l["disponivel"]]
    emprestados = [l for l in lista_livros if not l["disponivel"]]

    print(f"│  Total: {len(lista_livros)} livro(s)  │  "
          f"✔ Disponíveis: {len(disponiveis)}  │  "
          f"✘ Emprestados: {len(emprestados)}")
    separador()
    print(f"  {'ID':<5} {'Título':<34} {'Autor':<26} {'Gênero':<18} {'Status'}")
    separador()

    for livro in lista_livros:
        if livro["disponivel"]:
            status = "✔ Disponível"
        else:
            emp = buscar_emprestimo_por_livro(livro["id"])
            if emp:
                usuario = buscar_usuario_por_id(emp["usuario_id"])
                nome = usuario["nome"] if usuario else "Desconhecido"
                status = f"✘ Com: {nome}"
            else:
                status = "✘ Emprestado"

        print(f"  {livro['id']:<5} {livro['titulo']:<34} {livro['autor']:<26} "
              f"{livro['genero']:<18} {status}")
    print()


def cadastrar_usuario():
    global proximo_id_usuario
    print("\n┌─── Cadastrar Usuário ────────────────────────────────")
    nome  = input("│  Nome    : ").strip()
    email = input("│  E-mail  : ").strip()

    if not nome or not email:
        print("└  ✘  Nome e e-mail são obrigatórios.\n")
        return

    for usuario in lista_usuarios:
        if usuario["email"] == email:
            print("└  ✘  Já existe um usuário com esse e-mail.\n")
            return

    lista_usuarios.append({
        "id": proximo_id_usuario,
        "nome": nome,
        "email": email,
    })
    print(f"└  ✔  Usuário '{nome}' cadastrado com ID {proximo_id_usuario}.\n")
    proximo_id_usuario += 1


def realizar_emprestimo():
    global proximo_id_emprestimo

    print("\n┌─── Realizar Empréstimo ──────────────────────────────")

    disponiveis = [l for l in lista_livros if l["disponivel"]]
    if not disponiveis:
        print("│  ✘  Não há livros disponíveis no momento.\n└")
        return

    print("│  Livros disponíveis:")
    for l in disponiveis:
        print(f"│    [{l['id']}] {l['titulo']} — {l['autor']}")
    separador("·")

    try:
        livro_id   = int(input("│  ID do Livro   : "))
        usuario_id = int(input("│  ID do Usuário : "))
    except ValueError:
        print("└  ✘  IDs devem ser números inteiros.\n")
        return

    livro   = buscar_livro_por_id(livro_id)
    usuario = buscar_usuario_por_id(usuario_id)

    if livro is None:
        print("└  ✘  Livro não encontrado.\n")
        return
    if usuario is None:
        print("└  ✘  Usuário não encontrado.\n")
        return

    if livro["disponivel"]:
        livro["disponivel"] = False
        lista_emprestimos.append({
            "id": proximo_id_emprestimo,
            "livro_id": livro_id,
            "usuario_id": usuario_id,
        })
        proximo_id_emprestimo += 1
        print(f"└  ✔  '{livro['titulo']}' emprestado para {usuario['nome']}.\n")
    else:
        emp     = buscar_emprestimo_por_livro(livro_id)
        titular = buscar_usuario_por_id(emp["usuario_id"]) if emp else None
        nome    = titular["nome"] if titular else "outro usuário"
        print(f"└  ✘  Indisponível — '{livro['titulo']}' está com {nome}.\n")


def devolver_livro():
    print("\n┌─── Devolver Livro ───────────────────────────────────")

    emprestados = [l for l in lista_livros if not l["disponivel"]]
    if not emprestados:
        print("│  ✔  Nenhum livro está emprestado no momento.\n└")
        return

    print("│  Livros atualmente emprestados:")
    for l in emprestados:
        emp     = buscar_emprestimo_por_livro(l["id"])
        usuario = buscar_usuario_por_id(emp["usuario_id"]) if emp else None
        nome    = usuario["nome"] if usuario else "Desconhecido"
        print(f"│    [{l['id']}] {l['titulo']} — com: {nome}")
    separador("·")

    try:
        livro_id = int(input("│  ID do Livro : "))
    except ValueError:
        print("└  ✘  ID deve ser um número inteiro.\n")
        return

    livro = buscar_livro_por_id(livro_id)

    if livro is None:
        print("└  ✘  Livro não encontrado.\n")
        return

    if not livro["disponivel"]:
        livro["disponivel"] = True
        for i, emp in enumerate(lista_emprestimos):
            if emp["livro_id"] == livro_id:
                lista_emprestimos.pop(i)
                break
        print(f"└  ✔  '{livro['titulo']}' devolvido e já está disponível novamente.\n")
    else:
        print(f"└  ✘  '{livro['titulo']}' já está na biblioteca, não estava emprestado.\n")


def ver_emprestimos():
    print("\n┌─── Empréstimos Ativos ───────────────────────────────")

    if not lista_emprestimos:
        print("│  Nenhum empréstimo ativo no momento.\n└")
        return

    print(f"│  Total de empréstimos ativos: {len(lista_emprestimos)}")
    separador()
    print(f"  {'Emp.':<6} {'Livro':<34} {'Usuário':<22} {'E-mail'}")
    separador()

    for emp in lista_emprestimos:
        livro   = buscar_livro_por_id(emp["livro_id"])
        usuario = buscar_usuario_por_id(emp["usuario_id"])
        titulo  = livro["titulo"]   if livro   else "?"
        nome    = usuario["nome"]   if usuario else "?"
        email   = usuario["email"]  if usuario else "?"
        print(f"  {emp['id']:<6} {titulo:<34} {nome:<22} {email}")
    print()


def exibir_menu():
    print("╔═══════════════════════════════════════╗")
    print("║        SISTEMA DE BIBLIOTECA          ║")
    print("╠═══════════════════════════════════════╣")
    print("║  1 - Cadastrar Livro                  ║")
    print("║  2 - Listar Acervo Completo           ║")
    print("║  3 - Cadastrar Usuário                ║")
    print("║  4 - Realizar Empréstimo              ║")
    print("║  5 - Devolver Livro                   ║")
    print("║  6 - Ver Empréstimos Ativos           ║")
    print("║  0 - Sair                             ║")
    print("╚═══════════════════════════════════════╝")
    return input("  Escolha uma opção: ").strip()


def main():
    print("╔═══════════════════════════════════════╗")
    print("║     BEM-VINDO À BIBLIOTECA DIGITAL    ║")
    print("╚═══════════════════════════════════════╝")
    inicializar_dados()

    while True:
        opcao = exibir_menu()

        if opcao == "1":
            cadastrar_livro()
        elif opcao == "2":
            listar_livros()
        elif opcao == "3":
            cadastrar_usuario()
        elif opcao == "4":
            realizar_emprestimo()
        elif opcao == "5":
            devolver_livro()
        elif opcao == "6":
            ver_emprestimos()
        elif opcao == "0":
            print("\n  Encerrando o sistema. Até logo!\n")
            break
        else:
            print("  ✘  Opção inválida. Tente novamente.\n")


if __name__ == "__main__":
    main()
```
